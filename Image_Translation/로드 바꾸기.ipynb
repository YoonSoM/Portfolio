```python
# 경로는 각자 바꾸기

data_path = '/content/drive/MyDrive/팀플 1조 (객체인식) /data/training' 
os.listdir(data_path)
```

```python
json_path = []

# 이미지와 라벨 리스트에 담기
for label in os.listdir(data_path):
  sub_path = data_path + '/' + label + '/'

  for filename in os.listdir(sub_path):
    if filename[-4:] == 'json' :
      json_path.append(sub_path + filename)

len(json_path)
```

```python
# 라벨할 것들 정하기 (일단 전부다 넣음)

names = ['Cigar', 'Face', 'Leye', 'Mouth', 'Phone', 'Reye']
```

```python
image_width = 800
image_height = 1280

#json_path
for path in json_path :
  
  #json 파일 열고
  with open(path, 'r') as json_file:
    json_load_file = json.load(json_file)
    
    # txt에 라벨+좌표값 넣기
    with open(path[:-4]+'txt', 'w',encoding='UTF-8') as file:
      for idx, name in enumerate(names) :
        x1, y1, x2, y2 = json_load_file['ObjectInfo']['BoundingBox'][name]['Position'] # 객체 1개의 bbox 위치 좌표
        if x1 != x2:
          x = (x2-x1)/(2*image_width)
          y = (y2-y1)/(2*image_height)
          w = (x2-x1)/image_width
          h = (y2-y1)/image_height

          file.write(f'{idx} {x:f} {y:f} {w:f} {h:f}\n')
    file.close()
```
