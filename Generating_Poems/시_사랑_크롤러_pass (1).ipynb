```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re 
from tqdm import tqdm


# 본문, 타이틀, 저자  wr_id=26~ 266865, page = 1~6131 

# http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01&wr_id=266865&page=1
# http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01&page=1

# 'http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01' + '&wr_id=266865' + '&page=1'
```

```python
si = {
    'title' : [],
    'content' : [],
    'author' : []
}

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.0.0 Safari/537.36'}


for page in tqdm(range(1, 6)):
    url = 'http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01&page=' + str(page)
    resp = requests.get(url,headers = headers, verify = False)
    soup = BeautifulSoup(resp.text, 'lxml')
    links = soup.select('div.wr-subject a.item-subject')
    wr_id = [ re.findall('id=(.+)&',l['href'])[0] for l in links] # 여기까지는 ok

    for id in wr_id:
        try:
            url = 'http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01&' + 'wr_id=' + id + '&page=' + str(page)
            resp = requests.get(url,headers = headers, verify = False)
            soup = BeautifulSoup(resp.text, 'lxml')
            content = soup.select('div.view-content')[0]
            author = soup.select('div.panel-heading strong')[0].text
            title = soup.select('div.view-wrap h1')[0].text.strip()
            si['content'].append(content.text.strip().replace('\r', ' '))
            si['author'].append(author)
            si['title'].append(title)
        except Exception as e:
            print(f'page: {page}, id: {id}')
            print(e)
```

```python
import pandas as pd

df = pd.DataFrame(si)
df
```
